```python
import pandas as pd
```

# Rain data

```python
rain = pd.read_csv('data/rain_data_aus.csv',sep = ',')
rain.head(50)
```

## dropando colunas
 Amount of rain, Model estÃ£o vazando dados

```python
rain = rain.drop(['modelo_vigente','amountOfRain'],axis = 1)
```

# Wind

```python
wind = pd.concat([pd.read_csv('data/wind_table_01.csv',sep=',') for i in range(1,8)])
```

```python
wind.head(5)
```

# wind_rain

```python
wind_rain =pd.merge(rain,wind,on=['date','location'],how ='left')
wind_rain.head()
```

# Melhorando as colunas mais relevantes

## humidity_3pm

```python
humidity9am = wind_rain[wind_rain['humidity3pm'].isna()]['humidity9am']
wind_rain.loc[wind_rain['humidity3pm'].isna(),'humidity3pm'] = humidity9am
humidity = wind_rain[wind_rain['humidity3pm'].isna()]['humidity']
wind_rain.loc[wind_rain['humidity3pm'].isna(),'humidity3pm'] = humidity
```

```python
wind_rain[wind_rain['humidity3pm'].isna()]
```

## humidity

```python
humidity9am = wind_rain[wind_rain['humidity'].isna()]['humidity9am']
wind_rain.loc[wind_rain['humidity'].isna(),'humidity'] = humidity9am
```

# Pressure

## Pressure9am

```python
pressure3pm = wind_rain[wind_rain['pressure9am'].isna()]['pressure3pm']
wind_rain.loc[wind_rain['pressure9am'].isna(),'pressure9am'] = pressure3pm
```

## Pressure3pm

```python
pressure9am = wind_rain[wind_rain['pressure3pm'].isna()]['pressure9am']
wind_rain.loc[wind_rain['pressure3pm'].isna(),'pressure3pm'] = pressure9am
```

```python
wind_rain[wind_rain['pressure3pm'].isna()]
```

```python
wind_rain.dropna(inplace=True)
```
